```python
#now we have to check the ratio of poisonous mushrooms and eatable mushrooms so for that we need to plot a graph 
eatable_count = mushroom[mushroom['class']=='e']['class'].count()
poisonus_count = mushroom[mushroom['class']=='p']['class'].count()
p_e_ratio = poisonus_count/float(eatable_count)
print("\n'poisonus'-to 'eatable' ratio:{}\poisonus_count:{}, eatable_count: {}"
      .format(p_e_ratio.round(2),poisonus_count,eatable_count))


#need to check the balance between "e" and "p" hence
sns.set(style="ticks",color_codes=True)
plt.title("Balance Checking for input class in Mushroom Dataset",fontsize=15)
sns.countplot(x = mushroom["class"], data=mushroom)
```

```python
#'poisonus'-to 'eatable' ratio:0.93\poisonus_count:3916, eatable_count: 4208

#now we have to check the unique value of mushrooms hence
columns = mushroom.columns.values
#using for loop
for column in columns:
    print("{0}: {1}".format(column,mushroom[column].unique()))
```

```python
#now drop the veil-type as it has only one unique value no of features are now reduced to 21 columns
mushroom=mushroom.drop('veil-type',axis =1)
mushroom.shape
```

```python
#now use one-hot encoder to encode the mushrooms
mushroom_onehot = pd.get_dummies(mushroom)
#as we are dealing with the classification problem we can drop class of eatable mushroom and use only poisonus class 
mushroom_onehot = mushroom_onehot.drop(['class_e'],axis=1)


encoded= list(mushroom_onehot.columns[1:])
print("{} total feartures after one-hot encoding.".format(len(encoded)))
print(encoded)

mushroom_onehot.head(5)
```

```python
#important analysis for features
import numpy as np
feature_columns = mushroom.columns[1:]
sns.set(style='darkgrid', color_codes=True)
fig, axes = plt.subplots(nrows=3, ncols=7,figsize=(35,15))
mushroom['id']= np.arange(1,mushroom.shape[0] + 1)
#here using for loop
for f, ax in zip(feature_columns,axes.ravel()):
    mushroom.groupby(['class', f])['id'].count().unstack(f).plot(kind='bar', ax=ax, legend=True,grid=True,fontsize=18)
    ax.set_title(f, style='oblique',size=24)
    ax.set_title('\n', style='italic',size=18)
    ax.legend(fontsize=14)
    
mushroom=mushroom.drop("id",axis=1)
#here legend is used to easily identify the colours in the graph
```

```python
#now we have to calcutalte the correlation by pd.corr()
corr= mushroom_onehot.corr().loc[:,'class_p']
top_10_corr = corr.abs() . sort_values(ascending=False).head(n=11).iloc[1:]
print("Top_10 features to class_p correlation:","\n\n",top_10_corr)


top_10_corr_ratio = pd.DataFrame(index=range(2))
for feature in top_10_corr.index:
    feature_grouped = mushroom_onehot[['class_p',feature]].groupby([feature])
    top_10_corr_ratio.loc[:feature] = 100+feature_grouped.sum()/(poisonus_count)
    
    
print('\n\nTop-10 features-class_p poisonus mushrooms ratio:')
top_10_corr_ratio
```
